```r
ENSgenenames <- read.table(file = '/nfs/research1/marioni/Leah/scNMTseq/ModelSystems/pipeline_draft/evaluating/ENSgenenames.tsv', sep = '\t', header = FALSE)

getEnsID <- function(genename, ENSgenenames) {
    return(unlist(lapply(genename, function(gene) ENSgenenames$V1[which(ENSgenenames$V3 == gene)])))
}
```

```r
# Create seurat object
srat <- CreateSeuratObject(srat.processed@assays$RNA@counts, meta.data = srat.processed@meta.data)

mito.genes <- getEnsID(grep(pattern = "^mt-", x = ENSgenenames$V3, value = TRUE), ENSgenenames)
ribo.genes <- c(grep(pattern = "^Rpl", x = ENSgenenames$V3, value = TRUE),grep(pattern = "^Rps", x = ENSgenenames$V3, value = TRUE))
mito.genes <- mito.genes[mito.genes %in% rownames(srat)]
ribo.genes <- ribo.genes[ribo.genes %in% rownames(srat)]
srat[["percent.mt"]] <- PercentageFeatureSet(srat, features = mito.genes)
srat[["percent.ribo"]] <- PercentageFeatureSet(srat, features = ribo.genes)
```

```r
srat_WT <- subset(srat, subset = genotype == "WT")
srat_TKO <- subset(srat, subset = genotype == "TKO")
```

```r
options(repr.plot.width=10, repr.plot.height=5)

VlnPlot(srat_WT, features = c("nFeature_RNA", "nCount_RNA", "percent.mt",  "percent.ribo"), ncol = 4)
VlnPlot(srat_TKO, features = c("nFeature_RNA", "nCount_RNA", "percent.mt",  "percent.ribo"), ncol = 4)
```

```r
options(repr.plot.width=20, repr.plot.height=5)

for (s in c("srat_WT", "srat_TKO")) {

    toplot <- data.frame(
      nFeature_RNA = get(s)$nFeature_RNA,
      nCount_RNA = get(s)$nCount_RNA,
      percent.mt = get(s)$percent.mt,
      percent.ribo = get(s)$percent.ribo
    )

    p1 <- (ggplot(toplot, aes(x=nFeature_RNA)) +
        geom_histogram( color="#e9ecef", alpha=0.4, position = 'identity') +
        scale_fill_manual(values=c("#69b3a2")) +
        labs(fill="") +
        scale_x_continuous(trans='log2') +
        geom_vline(xintercept=675) +
        ggtitle(s) +
        theme_classic())

    p2 <- (ggplot(toplot, aes(x=nCount_RNA)) +
        geom_histogram( color="#e9ecef", alpha=0.4, position = 'identity') +
        scale_fill_manual(values=c("#69b3a2")) +
        labs(fill="") +
        scale_x_continuous(trans='log2') +
        geom_vline(xintercept=1250) +
        ggtitle(s) +
        theme_classic())

    p3 <- (ggplot(toplot, aes(x=percent.mt)) +
        geom_histogram( color="#e9ecef", alpha=0.4, position = 'identity') +
        scale_fill_manual(values=c("#69b3a2")) +
        labs(fill="") +
        #geom_vline(xintercept=10) +
        ggtitle(s) +
        theme_classic())

    p4 <- (ggplot(toplot, aes(x=percent.ribo)) +
        geom_histogram( color="#e9ecef", alpha=0.4, position = 'identity') +
        scale_fill_manual(values=c("#69b3a2")) +
        labs(fill="") +
        ggtitle(s) +
        theme_classic())

    grid.arrange(p1, p2, p3, p4, ncol=4)

}
```

```r
####################
## Load data sets ##
####################

message("Loading dataset")

srat.processed <- readRDS(paste0(srat_path))

srat <- CreateSeuratObject(srat.processed@assays$RNA@counts, meta.data = srat.processed@meta.data); rm(srat.processed)


########
## QC ##
########

mtx_sub_list <- list()
cell.info_sub_list <- list()
celltypes <- unique(srat@meta.data$genotype)
for (c in celltypes) {
    foo <- io$min_nFeature_RNA[["TettKOchimera"]][[c]]
    bar <- io$min_nCount_RNA[["TettKOchimera"]][[c]]
    tmp <- subset(srat, subset = nFeature_RNA > foo & nCount_RNA > bar, cells = rownames(srat@meta.data)[which(srat@meta.data$genotype == c)])
    cell.info_sub_list[[c]] <- tmp@meta.data
    cell.info_sub_list[[c]]$origrn <- rownames(cell.info_sub_list[[c]])
    mtx_sub_list[[c]] <- tmp@assays$RNA@counts
}
mtx <- do.call("cbind", mtx_sub_list)
rownames(mtx) <- gene.info$ens_id
cell.info <- do.call("rbind", cell.info_sub_list)
rownames(cell.info) <- cell.info$origrn

############
## Seurat ##
############

# Create seurat object

message("Creating seurat...")

srat <- CreateSeuratObject(mtx, meta.data = cell.info)


##########################
## SingleCellExperiment ##
##########################

# sce <- as.SingleCellExperiment(srat)

##########
## Save ##
##########

saveRDS(srat, "/nfs/research1/marioni/Leah/scNMTseq/ModelSystems/TettKOchimera/processed_files_Leah/seurat_QCed.rds")
fwrite(cell.info, "/nfs/research1/marioni/Leah/scNMTseq/ModelSystems/TettKOchimera/processed_files_Leah/sample_metadata_QCed.txt", quote=F, na="NA", sep="\t")
```
